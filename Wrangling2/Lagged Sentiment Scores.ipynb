# Create Lagged Sentiment Scores

This program will create a lagged variable that looks at the average sentiment 2 days, 14 days, and one month before an event occurs. At the end the data should have each date be unique, with a lagged variable cooresponding to the number of days lagged. 

```python
import pandas as pd
import numpy as np
import os
```

```python
merged_path = "/Users/laurieottehenning/Documents/Georgetown Data Science /Capstone/Merged"
os.chdir(merged_path)
es = pd.read_csv('Merged Events_Sentiment.csv')
es.groupby(['EventID', 'Date']).size().count()
```

```python
# Drop unnamed variable
es = es.drop('Unnamed: 0', 1)
```

```python
# View data
es = es.sort_values(by='Date', ascending=True)
es.sort_values(by='Date', ascending=True).iloc[1000:1010]
```

### Average Tone over the past n days

This will create a rolling sum of the tone of the articles over the past n days

```python
es['Date'] = pd.to_datetime(es['Date'])

df1 = es.groupby(['year', 'month']).apply(lambda x: x.set_index('Date').resample('1D').first())
```

```python
df1.head(4)
```

```python
lag30 = df1.groupby(level=0)['tone'].apply(lambda x: x.shift()
           .rolling(min_periods=1,window=30).mean()).reset_index(
    name='Value_Average_Past_30_days')
lag14 = df1.groupby(level=0)['tone'].apply(lambda x: x.shift()
           .rolling(min_periods=1,window=14).mean()).reset_index(
    name='Value_Average_Past_14_days')
lag7 = df1.groupby(level=0)['tone'].apply(lambda x: x.shift()
           .rolling(min_periods=1,window=7).mean()).reset_index(
    name='Value_Average_Past_7_days')
```

#### Create a time lagged variable
Each lagged variable will be lagged t-1

```python
lag30['lag_30'] = lag30['Value_Average_Past_30_days'].shift(-1)
lag30.sort_values(by='Date').iloc[500:505]
```

```python
lag14['lag_14'] = lag14['Value_Average_Past_14_days'].shift(-1)
lag14.sort_values(by='Date').iloc[500:505]
```

```python
lag7['lag_7'] = lag7['Value_Average_Past_7_days'].shift(-1)
lag7.sort_values(by='Date').iloc[500:505]
```

### Import Publication Rating data

The publication rating data includes:
   * A dummy if newspaper publication is based in US
   * A dummy if newspaper publication is based in the Middle East (Includes North Africa, Iran, Pakistan, and Afghanistan)
   * A dummy if newspaper is a UK publication
   * A dummy for other
   * A dummy for publications friendly to Assad (Russian, Iran, China, Hezbollah)
   

We will create lagged variable grouped around each of these categories. 

```python
# Import Publication Rating Dataset
publication = "/Users/laurieottehenning/Documents/Georgetown Data Science /Capstone/Publication list"
os.chdir(publication)

pub = pd.read_csv('Publication List_categorized.csv')
```

```python
# List Columns
pub.columns
```

```python
# Drop Columns
pub.drop('Index', axis=1, inplace=True)
pub.drop('Assignee', axis=1, inplace=True)
```

```python
# Fill in NaN with zeros
pub.fillna(value=0, inplace=True)
```

```python
# Create new variable with all of the publication values in one variable
def fillval(row):
    if row['US Publication'] == 1:
        return 'US'
    if row['Middle East (Include North Africa, Turkey, Iran, Afghanistan, Pakistan)'] ==1:
        return 'Middle East'
    if row['UK Publication'] ==1:
        return 'UK'
    if row['Other'] ==1:
        return 'Other'
pub['publication origin'] = pub.apply(lambda row: fillval(row),axis=1)
```

```python
def fillval2(row):
    if row['Friendly to Assad (Russia, Iran, China, Hezbollah)'] ==1:
        return 'Friendly to Assad'
    elif row['Friendly to Assad (Russia, Iran, China, Hezbollah)'] ==0:
        return 'Not Friendly to Assad'
pub['friendly publication to assad'] = pub.apply(lambda row: fillval2(row), axis=1)
```

#### Create lagged variables for Publication Origin

```python
# Group with Publication Origin column
df2 = pub_es.groupby(['publication origin']).apply(lambda x: x.set_index('Date').resample('1D').first())
```

```python
# Create lag variables
lagpub30 = df2.groupby(level=0)['tone'].apply(lambda x: x.shift()
           .rolling(min_periods=1,window=30).mean()).reset_index(
    name='Publication_Origin_Value_Average_Past_30_days')
lagpub14 = df2.groupby(level=0)['tone'].apply(lambda x: x.shift()
           .rolling(min_periods=1,window=14).mean()).reset_index(
    name='Publication_Origin_Value_Average_Past_14_days')
lagpub7 = df2.groupby(level=0)['tone'].apply(lambda x: x.shift()
           .rolling(min_periods=1,window=7).mean()).reset_index(
    name='Publication_Origin_Value_Average_Past_7_days')
```

```python
# Create shift the Publication origian average to t-1
lagpub30['lagpub_30'] = lagpub30['Publication_Origin_Value_Average_Past_30_days'].shift(-1)
lagpub30.sort_values(by='Date').iloc[500:505]
```

```python
# Create shift the Publication origian average to t-1
lagpub14['lagpub_14'] = lagpub14['Publication_Origin_Value_Average_Past_14_days'].shift(-1)
lagpub14.sort_values(by='Date').iloc[500:505]
```

```python
# Shift the Publication origin average to t-1
lagpub7['lagpub_7'] = lagpub7['Publication_Origin_Value_Average_Past_7_days'].shift(-1)
lagpub7.sort_values(by='Date').iloc[500:505]
```

## Merge
#### Merge rolling average with data set

```python
df3 = pub_es[['EventID', 'Date', 'FinalRating', 'publication', 'publication origin']].copy()
# df3 = df3.drop_duplicates('EventID', keep='False')
df3.groupby(['EventID', 'Date']).size().count()
```

```python
t1 = (pd.merge(df3, lag30, on=['Date'], how='left'))

lag14 = lag14[['Date', 'Value_Average_Past_14_days', 'lag_14']].copy()
t2 = (pd.merge(t1, lag14, on=['Date'], how='left'))

lag7 = lag7[['Date', 'Value_Average_Past_7_days', 'lag_7']].copy()
t3 = (pd.merge(t2, lag7, on=['Date'], how='left'))

lagpub30 = lagpub30[['Date', 'Publication_Origin_Value_Average_Past_30_days', 'lagpub_30', 
                     'publication origin']]
t4 = (pd.merge(t3, lagpub30, on=['Date', 'publication origin'], how='left'))

lagpub14 = lagpub14[['Date', 'Publication_Origin_Value_Average_Past_14_days', 'lagpub_14', 
                     'publication origin']]
t5 = (pd.merge(t4, lagpub14, on=['Date', 'publication origin'], how='left'))

lagpub7 = lagpub7[['Date', 'Publication_Origin_Value_Average_Past_7_days', 'lagpub_7', 
                   'publication origin']]
result = (pd.merge(t5, lagpub7, on=['Date', 'publication origin'], how='left'))
```

```python
result.columns
```

#### Publication Origin

The function <span style="color:blue">merge_pub</span> will reshape the lagpub columns by publication origin to a wide format (there will be a seperate column for each publication origin value). This will be merged back into the original data set. 

```python
# Reshape
def merge_pub(pub):
    test = result[['Date', 'publication origin', pub]].copy()
    test2 = pd.pivot_table(test, values=pub, index=['Date'], 
                           columns=['publication origin'])
    test2.rename(columns={'Middle East': 'Middle East '+ pub, 'Other': 'Other '+ pub,
                         'UK': 'UK '+ pub, 'US': 'US '+ pub}, inplace=True)
    return test2
t = merge_pub('lagpub_30')
t2 = merge_pub('lagpub_14')
t3 = merge_pub('lagpub_7')

z = (pd.merge(t, result, right_on=['Date'], left_index=True))
z2 = (pd.merge(t2, z, right_on=['Date'], left_index=True))
final = (pd.merge(t3, z2, right_on=['Date'], left_index=True))
final.columns
```
